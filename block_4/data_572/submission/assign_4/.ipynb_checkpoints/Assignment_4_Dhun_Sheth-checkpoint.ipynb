```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix, ConfusionMatrixDisplay,f1_score
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
print(X.shape, y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn import svm

# Initialize SVM classifier
"YOUR CODE"
svm_model = svm.SVC()

# grid search for best parameters
"YOUR CODE"
param_grid = {
    'C': [1, 5, 10],
    'gamma': [0.0001, 0.001, 0.1]
}

svm_grid = GridSearchCV(svm_model, param_grid, cv=5, scoring='accuracy')
svm_grid.fit(X_train, y_train)
svm_grid_predictions = svm_grid.predict(X_test)

# Fit the model for grid search
"YOUR CODE"
print(svm_grid.best_params_)

# Predict the model using the best parameters
"YOUR CODE"
print(classification_report(y_test, svm_grid_predictions ))

# plot confusin matrix
"YOUR CODE"
cm = confusion_matrix(y_test, svm_grid_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier,  plot_tree

# Initialize decision tree classifier
"YOUR CODE"
d_tree = DecisionTreeClassifier()

# grid search CV
"YOUR CODE"
param_grid = {
    'max_depth': [1,3,4,6,8],
    'min_samples_split': [2,3,5]
}

decisionTree_grid = GridSearchCV(d_tree, param_grid, cv=5, scoring='accuracy')
decisionTree_grid.fit(X_train, y_train)
decisionTree_grid_predictions = decisionTree_grid.predict(X_test)
best_decision_tree_params = decisionTree_grid.best_params_
print('Best parameters',decisionTree_grid.best_params_) 

# print how our model looks after hyper-parameter tuning
"YOUR CODE"
best_decision_tree = DecisionTreeClassifier(**best_decision_tree_params)
best_decision_tree.fit(X_train, y_train)

plt.figure(figsize=(12, 8))
plot_tree(best_decision_tree, filled=True, feature_names=data.feature_names, class_names=[str(i) for i in best_decision_tree.classes_], rounded=True)
plt.title("Best Decision Tree")
plt.show()

# print classification report 
print(classification_report(y_test, decisionTree_grid_predictions))

# plot confusion matrix
"YOUR CODE"
cm = confusion_matrix(y_test, decisionTree_grid_predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[str(i) for i in best_decision_tree.classes_])
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize gradient boosting classifier
"YOUR CODE"
gbc = GradientBoostingClassifier()

# param_grid for gradient boosting classifier
"YOUR CODE"
param_grid = {
    'learning_rate': [ 0.01, 0.1],
    'max_depth': [2, 3],
    'n_estimators': [300, 400]
}

# Predict the model using the best parameters
"YOUR CODE"
gbc_grid = GridSearchCV(gbc, param_grid, cv=5, scoring='accuracy')
gbc_grid.fit(X_train, y_train)
gbc_grid_predictions = gbc_grid.predict(X_test)

print(gbc_grid.best_params_) 

# print classification report 
print(classification_report(y_test, gbc_grid_predictions))

# plot confusin matrix
"YOUR CODE"
cm = confusion_matrix(y_test, gbc_grid_predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[str(i) for i in gbc_grid.classes_])
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize random forest classifier
"YOUR CODE"
random_forest = RandomForestClassifier()

# grid search for best parameters
"YOUR CODE"
param_grid = {
    'criterion': ['gini'],
    'n_estimators': [50, 100],
    'max_depth': [1,4,6],
    'min_samples_split': [2,3,5]
}

# Predict the model using the best parameters
"YOUR CODE"
randomforest_grid = GridSearchCV(random_forest, param_grid, cv=5, scoring='accuracy')
randomforest_grid.fit(X_train, y_train)
randomforest_grid_predictions = randomforest_grid.predict(X_test)

print(randomforest_grid.best_params_) 

# print classification report 
print(classification_report(y_test, randomforest_grid_predictions))

# plot confusin matrix
"YOUR CODE"
cm = confusion_matrix(y_test, randomforest_grid_predictions)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[str(i) for i in randomforest_grid.classes_])
disp.plot(cmap='Blues', values_format='d')
plt.title("Confusion Matrix")
plt.show()
```
