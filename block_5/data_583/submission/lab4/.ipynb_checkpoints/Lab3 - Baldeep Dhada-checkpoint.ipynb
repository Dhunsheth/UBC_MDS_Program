### Question 1:

#### (a)

```python
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('lab3/lab3data.csv', parse_dates=['date'])
data.head()
```

```python
# Split the data into two parts based on the year
transactions_2023 = data[data['date'].dt.year == 2023]
transactions_2024 = data[data['date'].dt.year == 2024]
```

```python
transactions_2023.head()
```

```python
transactions_2024.head()
```

#### (b)

```python
# Grouping the 2023 data on customer ID
customer_groups_2023 = transactions_2023.groupby('customer_id')

customers_2023 = customer_groups_2023.agg({
    'age': 'first',
    'sex': 'first',
    'date': len,
    'value': 'sum'
})

customers_2023.rename(columns={'date': 'count'}, inplace=True)
customers_2023
```

#### (c)

```python
# Grouping the 2024 data on customer ID
customer_groups_2024 = transactions_2024.groupby('customer_id')

customers_2024 = customer_groups_2024.agg({
    'age': 'first',
    'sex': 'first',
    'date': len,
    'value': 'sum'
})

customers_2024.rename(columns={'date': 'count'}, inplace=True)
customers_2024
```

#### (d)

```python
# Join the two dataframes
customers_joined = customers_2023.join(customers_2024, how='outer', lsuffix='_2023', rsuffix='_2024')
```

```python
# Fill missing values with 0
customers_joined = customers_joined.fillna({'count_2024': 0, 'value_2024': 0})

# Filling values for missing age and sex
customers_joined['age_2024'] = customers_joined['age_2023'] + 1
customers_joined['sex_2024'] = customers_joined['sex_2023']
```

```python
# Print the joined data
customers_joined
```

#### (e)

```python
import statsmodels.gam.api as smg
import statsmodels.api as sm

bs = smg.BSplines(x=customers_joined[['age_2023', 'count_2023', 'value_2023']],
                 df=[4,5,5],
                 degree=[3,3,3])

exog = pd.get_dummies(customers_joined['sex_2023'], drop_first=True)
exog = sm.add_constant(exog) 

# Fitting GAM model
gam1 = smg.GLMGam(endog = customers_joined['count_2024']>0,
                exog = exog,
                smoother = bs,
                family = sm.families.Binomial())

res1 = gam1.fit(maxiter = 5000)
res1.summary()
```

#### (f)

```python
import statsmodels.gam.api as smg
import statsmodels.api as sm
import pandas as pd

bs = smg.BSplines(x=customers_joined[['age_2023', 'count_2023', 'value_2023']],
                 df=[4,5,5],
                 degree=[3,3,3])

exog = pd.get_dummies(customers_joined['sex_2023'], drop_first=True)
exog = sm.add_constant(exog)

# Fitting second GAM to predict the value of purchases each customer makes in the first three months of 2024
gam2 = smg.GLMGam(endog = customers_joined['value_2024'],
                exog = exog,
                smoother = bs,
                family = sm.families.Gaussian())  # Using Gaussian distribution for continuous values

res2 = gam2.fit(maxiter = 5000)
res2.summary()
```

#### (g)

```python
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt
import numpy as np

# Fitting KDE for 2023 values
kde_value_2023 = KernelDensity(bandwidth = 10)
kde_value_2023.fit(transactions_2023['value'].values.reshape(-1, 1))

# Fitting KDE for 2024 values
kde_value_2024 = KernelDensity(bandwidth = 10)
kde_value_2024.fit(transactions_2024['value'].values.reshape(-1, 1))

#Plotting the distributions of purchase values in each of 2023 and 2024
xr = np.linspace(0, max(transactions_2023['value'].max(), transactions_2024['value'].max()), 1000).reshape(-1, 1)
plt.plot(xr, np.exp(kde_value_2023.score_samples(xr)), label = '2023')
plt.plot(xr, np.exp(kde_value_2024.score_samples(xr)), label = '2024')
plt.title('Purchase Value Distributions in 2023 and 2024')
plt.xlabel('Purchase Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Predicting the total purchase value for each customer in 2024
predicted_values = res2.predict()

# Fitting KDE for actual values
customer_value_2024 = KernelDensity(bandwidth = 10)
customer_value_2024.fit(customers_joined['value_2024'].values.reshape(-1, 1))

# Fitting KDE for predicted values
predicted_value_2024 = KernelDensity(bandwidth = 10)
predicted_value_2024.fit(predicted_values.reshape(-1, 1))

# Plotting the predicted and actual distributions of total purchase value by customer in 2024
xr = np.linspace(0, max(customers_joined['value_2024'].max(), predicted_values.max()), 1000).reshape(-1, 1)
plt.plot(xr, np.exp(customer_value_2024.score_samples(xr)), label = 'Actual')
plt.plot(xr, np.exp(predicted_value_2024.score_samples(xr)), label = 'Predicted')
plt.title('Predicted and Actual Distributions of Total Purchase Value by Customer in 2024')
plt.xlabel('Purchase Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```
