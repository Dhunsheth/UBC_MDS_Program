```python
import pandas as pd
import statsmodels.gam.api as smg
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy.stats import gaussian_kde
import numpy as np
from sklearn.neighbors import KernelDensity
```

## Question 1
### Part A

```python
data = pd.read_csv('lab3data.csv', parse_dates=['date'])
data_2023 = data[data['date'].dt.year == 2023]
data_2024 = data[data['date'].dt.year == 2024]
print(data_2023)
print(data_2024)
```

### Part B

```python
data_2023_grouped = data_2023.groupby(['customer_id', 'age', 'sex']).agg(total_value=('value', 'sum'), unique_purchases=('value', 'count'))
data_2023_grouped = data_2023_grouped.reset_index()
data_2023_grouped
```

### Part C

```python
data_2024_grouped = data_2024.groupby(['customer_id', 'age', 'sex']).agg(total_value=('value', 'sum'), unique_purchases=('value', 'count'))
data_2024_grouped = data_2024_grouped.reset_index()
data_2024_grouped
```

### Part D

```python
merged_df = pd.merge(data_2023_grouped, data_2024_grouped, on='customer_id', how='outer')
merged_df['age_y'] = merged_df['age_x']
merged_df['sex_y'] = merged_df['sex_x']
merged_df.fillna({'total_value_y':0, 'unique_purchases_y':0}, inplace=True)
merged_df['unique_purchases_y'] = merged_df['unique_purchases_y'].astype(int)
merged_df
```

### Part E

```python
plt.scatter(merged_df['age_x'].values, merged_df['unique_purchases_y'].values)
plt.xlabel('Age')  # Adds a label to the x-axis
plt.ylabel('# of Unique Purchases in 2024')  # Adds a label to the y-axis
plt.grid(True)
plt.show()
```

```python
plt.scatter(merged_df['total_value_x'].values, merged_df['unique_purchases_y'].values)
plt.xlabel('Total Value in 2023')  # Adds a label to the x-axis
plt.ylabel('# of Unique Purchases in 2024')  # Adds a label to the y-axis
plt.grid(True)
plt.show()
```

```python
plt.scatter(merged_df['unique_purchases_x'].values, merged_df['unique_purchases_y'].values)
plt.xlabel('2023 Purchases')  # Adds a label to the x-axis
plt.ylabel('# of Unique Purchases in 2024')  # Adds a label to the y-axis
plt.grid(True)
plt.show()
```

```python
endog = merged_df['unique_purchases_y']
exog = merged_df[['age_x','total_value_x', 'unique_purchases_x']]

bs = smg.BSplines(exog.values, df=[4,3,5], degree=[2,2,3])

gam_bs = smg.GLMGam(endog, sm.add_constant(exog), smoother=bs)
res_bs = gam_bs.fit()
print(res_bs.summary())
```

### Part F

```python
endog = merged_df['total_value_y']
exog = merged_df[['age_x','total_value_x', 'unique_purchases_x']]
exog_with_const = sm.add_constant(exog)

bs = smg.BSplines(exog.values, df=[3,3,5], degree=[2,2,3])

gam_bs = smg.GLMGam(endog, exog_with_const, smoother=bs)
res_bs = gam_bs.fit()
print(res_bs.summary())

predicted_values_2024 = res_bs.predict()
```

### Part G

```python

kde_2023 = KernelDensity(bandwidth=10)
kde_2023.fit(data_2023['value'].values.reshape(-1,1))

kde_2024 = KernelDensity(bandwidth=10)
kde_2024.fit(data_2024['value'].values.reshape(-1,1))

x = np.linspace(min(data_2023['value'].min(), data_2024['value'].min()), max(data_2023['value'].max(), data_2024['value'].max()), 1000).reshape(-1,1)

plt.plot(x, np.exp(kde_2023.score_samples(x)), label='2023 Purchase KDE', color='blue', linestyle='--')
plt.plot(x, np.exp(kde_2024.score_samples(x)), label='2024 Purchase KDE', color='orange', linestyle='-')
plt.xlabel('Purchase Value')
plt.legend()
plt.show()
```

```python
kde_2024 = KernelDensity(bandwidth=10)
kde_2024.fit(merged_df['total_value_y'].values.reshape(-1,1))

kde_2024_pred = KernelDensity(bandwidth=10)
kde_2024_pred.fit(predicted_values_2024.reshape(-1,1))

x = np.linspace(merged_df['total_value_y'].min(), merged_df['total_value_y'].max(), 1000).reshape(-1,1)
plt.plot(x, np.exp(kde_2024.score_samples(x)), label='2024 Total Value Observed KDE', color='r', linestyle='-')
plt.plot(x, np.exp(kde_2024_pred.score_samples(x)), label='2024 Total Value Predicted KDE', color='b', linestyle='-')
plt.xlabel('Total Purchase Value')
plt.legend()
plt.show()
```
