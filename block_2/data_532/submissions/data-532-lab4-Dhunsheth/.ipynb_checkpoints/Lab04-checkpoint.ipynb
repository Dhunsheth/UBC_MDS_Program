```python
### Your code goes here to do DFS on G1.

def dfs(v):
    """
    Does a depth-first-traversal starting at node v. Puts adjacent nodes in a stack and checks nodes in a stack based on LIFO against a visited set. 
    """
    stack_v = list(G1.adj[v])
    set_v.add(v)
    print(v)
    for i in range(len(stack_v)-1,0,-1):
        if stack_v[i] in set_v:
            stack_v.pop()
        else:
            dfs(stack_v[i])
    return set_v

set_v = set()
dfs(list(G1.nodes)[0])
print(set_v)
```

- Write a function `distance_BFS` that takes in a graph and two nodes, and uses breadth-first search (BFS) to compute the degrees of separation between the two nodes. Use Queue in the function (2 Marks)


```python
def distance_BFS(g, node1, node2):
    """
    Does a breadth-first search for node2, each node is inserted into a queue in a tuple - (value, distance from node1).
    When the search reaches node2, the distance is returned, else -1 is returned to represent not found. 
    """
    queue_list = []
    visited = set()
    queue_list.append((node1,0))
    visited.add(node1)

    while queue_list:
        node = queue_list[0][0]
        distance = queue_list[0][1]
        queue_list.pop(0)

        if node == node2:
            return distance
        for i in list(g.adj[node]):
            if i not in visited:
                queue_list.append((i,distance + 1))
                visited.add(i)
    return -1

distance_BFS(rj, 'nurse', 'mercutio')
```

## Knapsack Problem - 3 Marks

The knapsack problem is another classic dynamic programming exercise. The generalization of this problem is very old and comes in many variations, and there are actually multiple ways to tackle this problem aside from dynamic programming.

The problem at its core is one of combinatorial optimization. Given a set of items, each with a mass and a value, determine the collection of items that results in the highest possible value while not exceeding some limit on the total weight. The variation we’ll look at is commonly referred to as the 0–1 knapsack problem, which restricts the number of copies of each kind of item to 0 or 1. More formally, given a set of n items each with weight w_i and value v_i along with a maximum total weight W, our objective is:

max Σv_ix_i, where Σ w_ix_i≤W

* Either complete the code below or you can write you own function to solve the Knapsack problem using DP (2 marks)
* Add a markdown cell below to explain the intuition behind how you can solve the Knapsack problem using Dynamic Problem (1 marks)

```python
import numpy as np
def knapsack(W, wt, val):
    """
    Returns the max value of the 0,1 knapsack problem. Solved using dynamic programming and making table.  
    """
    v = {}
    for i in range(len(wt)):
        for w in range(W+1):
            if w == 0:
                v[i,w] = 0
            elif i == 0:
                if w <  wt[i]:
                    v[i,w] = 0
                else:
                    v[i,w] = val[i]
            elif w < wt[i]:
                v[i,w] = v[i-1,w]
            else:
                v[i,w] = max(v[i-1,w],v[i-1,w-wt[i]] + val[i])
    return v[len(wt)-1,W]

           
           

#Testing the function Knapsack
wt = list(np.random.randint(0, 10, 3))  
val = list(np.random.randint(0, 100, 3))  
W = 15
print(wt)
print(val)
knapsack(W, wt, val)
```

#### Knapsack Problem Intuition

To solve the knapsack problem using dynamic programming, I compute a table of choices where the columns are weight going from 0 to target weight, and the rows are each object. The formula for computing the table is as follows:
$$V[i,w] = max(V[i-1,w],V[i-1,w-wt[i]] + val[i])$$ 
Here i represents each object, w is the target weight, wt is the array of object weights and val is the array of object values. The table is calculated horizontally across all weight columns and then down considering each new object.

1) For the first row, if the target weight is less than object 1 weight, value is simply 0, else its the value of object 1. 
2) The max value for target weight W is found in the bottom right index of the table.
